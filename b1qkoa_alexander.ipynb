```python
#функция извлекает необходимые данные из структуры возвращаемой запросом на авто.ру
def extrakt_data(offers):
    rows = []    
    for offer in offers:
        new_data = {}
        try: new_data['bodyType'] = offer['vehicle_info']['configuration']['human_name']
        except: new_data['bodyType'] = None
            
        try: new_data['brand'] = offer['vehicle_info']['mark_info']['code']
        except: new_data['brand'] = None
            
        try: new_data['car_url'] = offer['additional_info']['mobile_autoservices_url']
        except: new_data['car_url'] = None
            
        try: new_data['color'] = offer['color_hex']
        except: new_data['color'] = None
            
        try: new_data['complectation_dict'] = offer['vehicle_info']['complectation']
        except:  new_data['complectation_dict'] = None
            
        try:  new_data['description'] = offer['description']
        except:  new_data['description'] = None
            
        try:  new_data['engineDisplacement'] = offer['vehicle_info']['tech_param']['displacement']/1000
        except:  new_data['engineDisplacement'] = None
            
        try:  new_data['enginePower'] = offer['vehicle_info']['tech_param']['power']
        except:  new_data['enginePower'] = None
            
        try:  new_data['equipment_dict'] = offer['vehicle_info']['equipment']
        except:  new_data['equipment_dict'] = None
            
        try:  new_data['fuelType'] = offer['lk_summary'].split()[-1]
        except:  new_data['fuelType'] = None
            
        #new_data['image'] = offer['vehicle_info']['mark_info']['code']
        
        try:  new_data['mileage'] = offer['state']['mileage']
        except:  new_data['mileage'] = None
            
        try:   new_data['modelDate'] = offer['vehicle_info']['super_gen']['year_from']
        except:  new_data['modelDate'] = None
            
        try:   new_data['model_info'] = offer['vehicle_info']['model_info']
        except:  new_data['model_info'] = None
            
        try:   new_data['model_name'] = offer['vehicle_info']['model_info']['name']
        except:  new_data['model_name'] = None
            
        try:   new_data['name'] = offer['vehicle_info']['tech_param']['human_name']
        except:  new_data['name'] = None
            
        try:   new_data['numberOfDoors'] = offer['vehicle_info']['configuration']['doors_count']
        except:  new_data['numberOfDoors'] = None
            
        try:   new_data['parsing_unixtime'] = offer['price_history']['[0]']['create_timestamp']
        except:  new_data['parsing_unixtime'] = None
            
        try:   new_data['priceCurrency'] = offer['price_info']['currency']
        except:  new_data['priceCurrency'] = None
            
        try:   new_data['productionDate'] = offer['documents']['year']
        except:  new_data['productionDate'] = None
            
        try:   new_data['sell_id'] = offer['id']
        except:  new_data['sell_id'] = None
            
        try:   new_data['super_gen'] = offer['vehicle_info']['super_gen']
        except:  new_data['super_gen'] = None
            
       # try:   new_data['vehicleConfiguration'] = offer['vehicle_info']['configuration']
       #except:  new_data['vehicleConfiguration'] = None
            
        try:   new_data['vehicleTransmission'] = offer['vehicle_info']['tech_param']['transmission']
        except:  new_data['vehicleTransmission'] = None
            
        try:   new_data['vendor'] = offer['vehicle_info']['vendor']
        except:  new_data['vendor'] = None
            
        try:   new_data['Владельцы'] = offer['documents']['owners_number']
        except:  new_data['Владельцы'] = None
            
        try:   new_data['Владение'] = offer['documents']['purchase_date']
        except:  new_data['Владение'] = None
            
        try:   new_data['ПТС'] = offer['documents']['pts']
        except:  new_data['ПТС'] = None
            
        try:   new_data['Привод'] = offer['lk_summary'].split(', ')[-2]
        except:  new_data['Привод'] = None
            
        try:   new_data['Руль'] = offer['vehicle_info']['steering_wheel']
        except:  new_data['Руль'] = None
            
        #try:   new_data['Состояние'] = offer['description']
        #except:  new_data['Состояние'] = None
            
        try:   new_data['Таможня'] = offer['documents']['custom_cleared']
        except:  new_data['Таможня'] = None
            
        try: new_data['price'] = offer['price_info']['RUR']
        except: new_data['price'] = None
            
        try: new_data['hidden'] =  offer['additional_info']['hidden']
        except: new_data['hidden'] = None
            
        try: new_data['start_date'] =  offer['additional_info']['hot_info']['start_time']
        except: new_data['start_date'] = None
            
        rows.append(new_data)  
    
    return rows
```

```python
#запросы с кагла не работают
# выгрузка за последний год, по 2000 страниц с обьявлениями по каждой марке из теста

offers = []
new_train = pd.DataFrame()
n = 0
#установим параметры запроса
mark = 'BMW'  # марка
#model = 'M6'  # модель
geo_radius = 200  # радиус вокруг москвы = 200 км
geo_id = [213]  # Москва
top_days = "360"  # кол-во дней от текущей даты назад за которые запрашиваем объявления 
for model in models:
    for page in range(1, 2000):
        param = {'catalog_filter' : [{"mark": model}],
             'section': "used", #бу
             'category': "cars",
             'sort': "fresh_relevance_1-desc",
             "top_days":top_days,
             "geo_radius":geo_radius,
             "geo_id": geo_id,
             'page': page
            }
        response = requests.post(url, json=param, headers = HEADERS)
        try:
            data = response.json()
        except:
            print('response.json() end')
            break
            
        if len(data['offers']) == 0:
            print(model, ' end')
            break   
            
        try:
            offers.extend(data['offers'])  
        except:
            print('offers.extend(data[offers]) end')
            break
        n += len(data['offers'])        
        rows = extrakt_data(offers)
        offers = []        
        df = pd.DataFrame(rows)
        new_train = pd.concat([new_train, df])  
        
        if page % 500 == 0:
            print(model,' ',page, "status_code: ",response.status_code, len(data['offers']))
    
print(n)

new_train.to_csv('new_train8.csv', index=False) # сохраняем выгрузку
```

```python
#train1['Владение'].unique()
#сколько прошло месяцев с даты покупки
def calc_delta_ouned(x):
    if x['year'] > 0:
        now = datetime.now()
        return (now.year - x['year'])*12 - x['month'] + now.month
    else: return 0
```

```python
temp = pd.DataFrame(train1['Владение'])
pattern = re.compile('\d{4}')
#разобьем строку на год и месяц
temp['year'] = temp['Владение'].apply(lambda x:str(pattern.findall(str(x)))[2:-2])
pattern = re.compile('(\d{1,2})\}$')
temp['month'] = temp['Владение'].apply(lambda x:str(pattern.findall(str(x)))[2:-2])
#приведем к коректному виду
temp['year'] = temp['year'].map(lambda x: int(x) if len(x)==4 else 0)
temp['month'] = temp['month'].map(lambda x: int(x) if len(x)>0 else 0)
#посчитаем сколько прошло месяцев
temp['delta_own'] = temp.apply(lambda x: calc_delta_ouned(x), axis=1)
train1['ownership'] = temp['delta_own']


temp = pd.DataFrame(test['Владение'])
pattern = re.compile('^\d{1,2}')
#разобьем строку на год и месяц
temp['year'] = temp['Владение'].apply(lambda x:str(pattern.findall(str(x)))[2:-2])
pattern = re.compile('и\s(\d{1,2})\sм')
temp['month'] = temp['Владение'].apply(lambda x:str(pattern.findall(str(x)))[2:-2])
#приведем к коректному виду
temp['year'] = temp['year'].map(lambda x: int(x) if len(x)>0 else 0)
temp['month'] = temp['month'].map(lambda x: int(x) if len(x)>0 else 0)
#посчитаем сколько прошло месяцев
temp['delta_own'] = temp.apply(lambda x: x['year']*12 + x['month'], axis=1)
test['ownership'] = temp['delta_own']
```
